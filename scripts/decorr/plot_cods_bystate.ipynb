```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
import pickle
```

```python
from bci_plot.utils import data_util
from bci_plot.metadata import decorr_info
from bci_plot.gen_data.decorr import gen_decorr_cod
from bci_plot.gen_fig.decorr import gen_decorr_fig_rendered
```

```python
all_cods = []
for session in decorr_info.decorr_sessions:
    try:
        fname = f'../../data/decorr/{session}_decorr_bystate.pickle'
        with open(fname, 'rb') as f:
            all_cods.append(pickle.load(f))
    except Exception as e:
        print(e)
        pass
all_corrs = []
for session in decorr_info.decorr_sessions:
    try:
        fname = f'../../data/decorr/{session}_corrs.pickle'
        with open(fname, 'rb') as f:
            all_corrs.append(pickle.load(f))
    except Exception as e:
        print(e)
        pass
```

```python
# train on each state, show cod for each state
titles = ['H1 Day T+0 CL_1', 'H2 Day T+0 CL_1', 'H2 Day T+0 CL_1', 'H4 Day T+0 CL_1', 'S2 Day T+0 CL_1', 'S2 Day T+0 CL_2', 'S2 Day T+2 CL_2', 'S2 Day T+2 CL_5', 'S2 Day T+4 CL_1']
fig = gen_decorr_fig_rendered.gen_decorr_fig2(all_cods, titles=titles, n_cols=3)
plt.savefig('../../figs/decorr/cods_bystate.pdf')
```

```python
cods_max = []
for i, cods in enumerate(all_cods):
    data = np.array([[[item[fold][state] for item in cods['gaze_from_h']['linear_by_state']] for state in [0, 1, 2, 3]] for fold in range(5)])
    cods_max.append(data.mean(0).max(0).max(0))
cods_max = np.array(cods_max)
```
